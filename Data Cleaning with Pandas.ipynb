```python
#Import data
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

flights_data = pd.read_csv('flights.csv')
flights_data.head(10)
weather_data_pd = pd.read_csv('weather.csv')
weather_data_np = weather_data_pd.to_numpy()
```

```python
#flights table
df = pd.read_csv('flights.csv')
df
```

```python
#weather table
dw = pd.read_csv('weather.csv')
dw
```

```python
#Question 1 How many flights were there from JFK to SLC? Int
jfk_slc = df[(df['origin'] == 'JFK') & (df['dest'] == 'SLC')]
q_one = len(jfk_slc)
print(f"Number of flights ftom JFK to SLC: {q_one}")
```

```python
#Question 2 How many airlines fly to SLC? Should be int
q_two = df[df['dest'] == 'SLC']['carrier'].nunique()
print(f"The number of Airlines flying to SLC: {q_two}")
```

```python
#Question 3 What is the average arrival delay for flights to RDU? float
q_three = df[df['dest'] == 'RDU']['arr_delay'].mean()
print(f"Average arrival delay for flights to RDU: {q_three:.2f} minutes")
```

```python
#Question 4 What proportion of flights to SEA come from the two NYC airports (LGA and JFK)?  float
flights_sea = df[df['dest'] == 'SEA']

nyc_to_sea = flights_sea[flights_sea['origin'].isin(['LGA', 'JFK'])]

q_four = len(nyc_to_sea) / len(flights_sea)
print(f"Proportion of flights to SEA from LGA and JFK: {q_four:.2f}")
```

```python
#Question 5 Which date has the largest average depature delay? Pd slice with date and float
#please make date a column. Preferred format is 2013/1/1 (y/m/d)

#create a new column?
df['date'] = pd.to_datetime(df[['year', 'month', 'day']]).dt.strftime('%Y/%m/%d')


avg_delay_date = df.groupby('date')['dep_delay'].mean()

q_five_delay_date = avg_delay_date.idxmax()
q_five_delay = avg_delay_date.max()

print(f"The date with the largest average departure delay is {q_five_delay_date} with an average delay of {q_five_delay:.2f} minutes.")
```

```python
#Question 6 Which date has the largest average arrival delay? pd slice with date and float
import pandas as pd

df['date'] = pd.to_datetime(df[['year', 'month', 'day']]).dt.strftime('%Y/%m/%d')

avg_delay_date2 = df.groupby('date')['arr_delay'].mean()

q_six_delay_date = avg_delay_date2.idxmax()
q_six_arr_delay = avg_delay_date2.max()

print(f"The date with the largest average arrival delay is {q_six_delay_date} with an average delay of {q_six_arr_delay:.2f} minutes.")
```

```python
#Question 7 Which flight departing LGA or JFK in 2013 flew the fastest? pd slice with tailnumber and speed
#speed = distance/airtime

flights_2013 = df[(df['origin'].isin(['LGA', 'JFK'])) & (df['year'] == 2013)]

flights_2013['speed'] = flights_2013['distance'] / flights_2013['air_time']

fastest_flight = flights_2013.loc[flights_2013['speed'].idxmax(), ['tailnum', 'speed']]

print(f"The fastest flight departing from LGA or JFK in 2013 was {fastest_flight['tailnum']} with a speed of {fastest_flight['speed']:.2f} miles per minute.")
```

```python
#Use weather_data_np
#Question 9 How many observations were made in Feburary?Int

month = weather_data_np[:, 3]
print(np.count_nonzero(month == 2.0))
```

```python
#Question 10 What was the mean for humidity in February? Float
humidity = weather_data_np[:,(3,8)]
mask = (humidity[:, 0] == 2.0)
feb = humidity[mask, :]
average_humidity = np.mean(feb[:, 1])
print(average_humidity)
```

```python
#Question 11 What was the std for humidity in February? Float
humidity = weather_data_np[:,(3,8)]
mask = (humidity[:, 0] == 2.0)
feb = humidity[mask, :]
std_humidity = np.std(feb[:, 1])
print(std_humidity)
```
